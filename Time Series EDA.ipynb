## Time Resampling

```python
df
```

```python
# will produce the min of all columns at each year end
df.resample(rule='A').min()
```

```python
# will produce the max of all columns at each year end
# rule A for each year end
 
df.resample(rule='A').max()
```

```python
df.resample(rule='A').max()['Open'].plot()
```

```python
# will produce the max of all columns quaterly each year
# rule QS produce quaterly
#https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
 
df.resample(rule='QS').max()
```

```python
df.resample(rule='QS').max()['Open'].plot()
```

```python
# will produce the max of all columns quaterly each year based on business
# rule BQS produce business quaters

#https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
 
df.resample(rule='BQS').max()
```

```python
# plotting 

df['Open'].resample(rule='BQS').mean().plot(kind='bar')
```
